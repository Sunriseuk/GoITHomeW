И помните, что переменные не должны совпадать с зарезервированными словами языка. Иначе компилятор не сможет понять оператор это или переменная.

С помощью операторов мы выполняем основные арифметические операции. Символы, с помощью которых выполняются операции в языке Python, называются операторами. Основные математические операторы перечислены ниже в таблице.
Символ 	Операция
+ 	Сложение
- 	Вычитание

a = 5

Порядок операторов в выражении

Математические операции выполняются по очереди, в зависимости от их приоритета (если не задать другую очередность с помощью скобок). Умножение и деление имеют более высокий приоритет, чем сложение и вычитание.

ratio = 10
result = 8 * (ratio + 5) - ratio ** 2

Выражением называется совокупность переменных, операций, имен функций, скобок, которая может быть вычислена в соответствии с синтаксисом Python.

Например, 8 * (ratio + 5) - ratio**2 — это выражение, 8 * (ratio + 5) — выражение, ratio**2 — выражение и т.д.

Если возникает двоякая ситуация при чтении выражения, старайтесь их группировать вместе с помощью скобок. Возьмем, например, возведение в степень отрицательного числа -3 и возведем его в квадрат. На бумаге мы поступили бы именно так:

-3 ** 2  # -9

Мы получим -9, но это скорее всего не тот результат, что вы хотели. А дело в том, что в Python приоритет операции возведения в степень выше отрицания. Поэтому оборачиваем -3 в скобки и получаем ожидаемый результат.

(-3) ** 2  # 9

Вернёмся к предыдущей задаче. В переменной rate у нас находится тариф за электроэнергию — 1.68. Но теперь нам необходимо вести расчет дневного и ночного тарифа. Ночной тариф рассчитывается в половину дневного, т.е. rate надо поделить на 2. В переменной value_day — дневные показания счетчика, а в переменной value_night — ночные. Опять же задайте для них разумное целое число киловатт. Рассчитайте и поместите в переменную payment счет за электричество, но уже с учетом дневного и ночного тарифов.

```python
rate = 1.68
value_day = 50
value_night = 100
payment = value_day * rate + value_night * (rate / 2)
```

Комментарии

Комментарии в коде программы крайне полезные в любом языке программирования. Пока рассматриваемые примеры программы достаточно простые, но по мере роста сложности кода в программы добавляют комментарии, описывающие общий подход к решаемой задаче, — своего рода заметки, написанные на понятном языке. Эта информация бывает крайне полезной, если приходится возвращаться к своей программе через долгое время или разбираться в коде, написанном другим разработчиком. При выполнении программы компьютер игнорирует все комментарии в коде, они присутствуют там исключительно для человека.

Для Python комментарий начинается с символа «решетки» (#). Интерпретатор Python игнорирует все символы, следующие в коде после # до конца строки. Пример:

ratio = 10  # коэффициент коррекции
result = 8 * (ratio + 5) - ratio ** 2

Опять вернёмся к предыдущей задаче. Добавим перед вычислением переменной payment поясняющий комментарий. Закомментируйте текст 'Payment for electricity for the current month', чтобы код стал рабочим.

```python
rate = 1.68
value_day = 358
value_night = 103
# Payment for electricity for the current month
payment = rate * value_day + rate / 2 * value_night
```

От обычной строки f-строка отличается тем, что в начале строки мы ставим символ f. Интерпретатор понимает, что внутри такой строки он встретит символы фигурных скобок { }, внутри которых содержится выражение, его необходимо выполнить и результат подставить в строку.

first = 'Hello'
second = 'world'
sentence = f"{first} {second}!"  # Hello world!

Типы данных. Числа

Пришло время чисел. В программировании они используются для ведения счета в играх, представления данных в визуализациях, хранения информации в веб-приложениях и т.д. В Python числовые данные разделены на несколько категорий по тому, как они используются. Меньше всего проблем для программиста приносит работа с целыми числами. Мы уже познакомились с основными операциями над числами в одном из первых разделов.

В Python числа, имеющие дробную часть, называются вещественными (или как принято говорить «числами с плавающей точкой»). Обычно мы их используем и не особо задумываемся о поведении этих чисел, но бывают ситуации, когда числа могут вести себя неожиданно для нас

a = 0.2 + 0.1  # 0.30000000000000004

Данная ситуация может произойти, независимо от языка программирования, и связана она с внутренним представлением чисел в компьютерах. Поэтому сравнение вещественных чисел возможно только с определенной точностью, например, до шестого знака после запятой. Это когда сравнивают не два числа, а их разность с точностью 0.000001 - меньше она точности или нет. Если при вычислениях точность становится важной, то стоит обратить внимание на модуль decimal (https://docs.python.org/3/library/decimal.html)

Комплексные числа

В Python встроены также и комплексные числа. Представление таких чисел происходит в виде пары значений: действительной и мнимой части, которые разделяются символом операции сложения. В завершении мнимой составной части комплексного числа обязательно ставят букву j. В следующем примере мы создаем комплексную переменную:

complex = 3.14 + 1j
print(complex)  # (3.14+1j)

Чтобы получить действительную и мнимую часть чисел, отдельно необходимо использовать свойства real и image:

complex = 3.14 + 1j
print(complex.real)  # 3.14
print(complex.imag)  # 1.0

Основная область применения комплексных чисел — это научные расчеты в математике, квантовой механике, электротехнике и т.п.

Необходимо выполнить сложение двух комплексных чисел (напомним, что комплексные числа поддерживают не все математические операции и не все операции сравнения).

Первая переменная a равна -2 + 3j и вторая переменная b равна 4 + 2.1j. Результат необходимо поместить в переменную result.

```python
a = -2 + 3j
b = 4 + 2.1j
result = a + b
```

Модуль math

Для работы с числами в Python существует один из наиважнейших модулей math (для комплексных чисел существует отдельный модуль cmath). Этот модуль предоставляет обширный функционал для работы с числовыми данными.

Чтобы добавить его в свою программу, необходимо выполнить импорт этого модуля с помощью ключевого слова import Например, чтобы извлечь квадратный корень из числа, надо использовать метод sqrt модуля math

import math

a = math.sqrt(100)  # 10.0

Подробно работа с модулями рассматривается дальше по курсу, пока мы только пользуемся готовыми для нас методами.

Необходимо вычислить корни квадратного уравнения.

a · x2 + b · x + c = 0

Задайте переменные коэффициенты a, b, c со значениями -2, 7, -6 соответственно

Дискриминант уравнения поместите в переменную D

D = b2 - 4 · a · c

Корни уравнения поместите в переменные x1 и x2

x1 = (-b + D0.5) / (2 · a)

x2 = (-b - D0.5) / (2 · a)

```python
import math

a = -2
b = 7
c = -6
D = pow(b, 2) - 4 * a * c
x1 = (-b + math.sqrt(D))/(2 * a)
x2 = (-b - math.sqrt(D))/(2 * a)
```
