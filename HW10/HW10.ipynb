```python
from collections import UserDict


class AdressBook(UserDict):

    def add_record(self, record):
        self.data[record.name.value] = record


class Record():
    def __init__(self, name, phone=None):
        self.name = Name(name)
        if phone:
            self.phones = [Phone(phone)]
        else:
            self.phones = []

    def add_phone(self, phone):
        self.phones.append(Phone(phone))

    def delete_phone(self, phone):
        phone_obj = Phone(phone)
        if phone_obj in self.phones:
            self.phone.remove(phone_obj)

    def change_phone(self, old_phone, new_phone):
        phone_obj = Phone(old_phone)
        if phone_obj in self.phones:
            self.phones[self.phones.index(phone_obj)] = Phone(new_phone)


class Field():
    def __init__(self, value):
        self.value = value


class Name(Field):
    pass


class Phone(Field):
    pass
```

```python
from adressbook import *

dict_telephones = AdressBook()


def input_error(func):
    """Обробляемо помилки, якіх міг припуститись користувач"""
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except KeyError:
            return "Ви ввели не вірне ім'я"
        except TypeError:
            return "Ви ввели не вірний форат команди"
        except IndexError:
            return "Введіть ім'я та телефон"
        except ValueError as e:
            return e.args[0]
        except Exception as e:
            return e.args

    return wrapper


@input_error
def answer_hello():
    """Привітаємо користувача"""
    return 'How can I help you?'


@input_error
def answer_add(name_telephone):
    """Додаємо телефон до словника"""
    lst_name_telephone = name_telephone.strip().split()
    name = lst_name_telephone[0]
    telephone = lst_name_telephone[1]

    if name in dict_telephones:
        record = dict_telephones.data[name]
        record.add_phone(telephone)
    else:
        record = Record(name, telephone)
        dict_telephones.add_record(record)

    return f'Для {name} записан телефон {telephone}'


@input_error
def answer_change(name_telephone):
    """Змінюємо телефон в словнику"""
    lst_name_telephone = name_telephone.strip().split()
    name = lst_name_telephone[0]
    old_telephone = lst_name_telephone[1]
    telephone = lst_name_telephone[2]
    record = dict_telephones.data[name]
    record.change_phone(old_telephone, telephone)
    return f'Для {name} змінено телефон {old_telephone} на {telephone}'


@input_error
def answer_phone(name):
    """Повертаємо телефон за ім'ям"""
    record = dict_telephones.data[name]
    return ', '.join([phone.value for phone in record.phones])


@input_error
def answer_showall():
    """Виводимо всі телефони"""
    text_phones = ''
    for name, record in dict_telephones.items():
        text_phones += name + ' ' + \
            ', '.join([phone.value for phone in record.phones]) + '\n'
    return text_phones

#    return '\n'.join([f'{name} {lambda record: "".join(record.phones)}' for name, record in dict_telephones.items()])


@input_error
def answer_exit():
    return 'Good bye!'


@input_error
def command_error():
    return 'Ви ввели не вірну команду'


"""Створимо словник команд для пошуку потрібних методів"""
DICT_COMMANDS = {
    'hello': answer_hello,
    'add': answer_add,
    'change': answer_change,
    'phone': answer_phone,
    'show all': answer_showall,
    'exit': answer_exit,
    'good bye': answer_exit,
    'close': answer_exit
}


def get_answer_function(answer):
    """Повертаємо потрибний метод для команди користувача"""
    return DICT_COMMANDS.get(answer, command_error)


@input_error
def run_command(user_command):
    command = user_command
    params = ''
    for key in DICT_COMMANDS:
        if user_command.lower().startswith(key):
            command = key
            params = user_command[len(command):]
            break
    if params:
        return get_answer_function(command)(params.strip())
    else:
        return get_answer_function(command)()


def main():
    while True:
        user_command = input('Введіть команду для бота: ')
        answer = run_command(user_command.strip())
        print(answer)
        if answer == 'Good bye!':
            break


if __name__ == "__main__":
    main()
```

Переменные класса и объекта

Существует два типа полей: переменные класса и переменные объекта. Как видно из названия они различаются тем, что принадлежат или классу или объекту соответственно.

Переменные класса — доступ к ним имеют все экземпляры этого класса. Переменная класса существует только одна, и любой из объектов, когда изменяет переменную класса, изменяет ее для всех остальных экземпляров этого же класса.

Переменные объекта — принадлежат каждому отдельному экземпляру класса. В этом случае у каждого объекта есть своя собственная копия поля, т.е. не разделяемая и никоим образом не связанная с другими такими же полями в других экземплярах.

Наследование

Основная сила объектно-ориентированного программирования заключается именно в возможности наследования классов. Наследование дает возможность создавать новые классы, содержащие атрибуты родительских классов.

class Human:
    name = ''

    def voice(self):
        print(f"Hello! My name is {self.name}")


class Developer(Human):
    field_description = "My Programming language"
    language = ""

    def make_some_code(self):
        return f"{self.field_description} is {self.value}"


class PythonDeveloper(Developer):
    value = "Python"


class JSDeveloper(Developer):
    value = "JavaScript"


p_dev = PythonDeveloper()
p_dev.name = 'Bob'
p_dev.voice()  # Hello! My name is Bob
p_dev.make_some_code()  # My Programming language is Python

js_dev = JSDeveloper()
js_dev.make_some_code()  # My Programming language is JavaScript

В этом примере мы создали родительский класс Human, который определил, что у всех есть имя и метод voice.

Далее мы расширили функционал класса Human и создали класс Developer, который наследуется от Human и добавляет ему поля field_description и language, и метод make_some_code. И чтобы реализовать функционал разработчика на конкретном языке, мы сделали два класса: PythonDeveloper и JSDeveloper. Теперь, если мы захотим добавить функционал во все дочерние для Human классы, то для этого можно добавить нужные атрибуты в Human, и они автоматически появятся и в PythonDeveloper, и JSDeveloper. Если надо расширить функционал всех разработчиков, то можно добавить нужные атрибуты в класс Developer. И если надо добавить особого поведения классу разработчиков на конкретном языке, то можно добавить или изменить атрибуты классов PythonDeveloper или JSDeveloper.

Таким образом, в коде должно быть только одно место, где определено поведение объекта. И если нам надо получить другой объект, у которого есть это поведение и какие-то свои особенности, мы можем унаследоваться от класса с нужными нам общими атрибутами и добавить уникальных.

Этот подход позволяет писать меньше кода и структурировать данные, создавая модели реальных объектов с их характеристиками (полями) и поведением (методами).

Имя super соответствует распространенной терминологии: класс-родитель называется суперклассом, а класс-потомок — субклассом. Значит, Animal — это суперкласс, а Cat — субкласс.

Экземпляры как атрибуты класса

При моделировании явлений реального мира классы со временем дополняются все большим количеством подробностей. Списки атрибутов и методов увеличиваются, и через некоторое время классы становятся длинными и трудно поддерживаемыми в разработке. В такой ситуации часть одного класса можно выделить в виде отдельного класса. Код большого класса разбивается на меньшие классы, которые работают во взаимодействии друг с другом.

Например, при дальнейшей доработке класса Cat может оказаться, что в нем стало слишком много атрибутов и методов, относящихся к хозяину кота. В таком случае все, что относится к владельцу кота, можно переместить в отдельный класс с именем Owner. Затем экземпляр Owner становится атрибутом класса Cat:

class Animal:
    def __init__(self, nickname, weight):
        self.nickname = nickname
        self.weight = weight

    def say(self):
        pass

    def change_weight(self, weight):
        self.weight = weight


class Owner:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address


class Cat(Animal):
    def __init__(self, nickname, weight, breed, owner):
        self.breed = breed
        self.owner = owner
        super().__init__(nickname, weight)

    def say(self):
        return "Meow"


owner = Owner("Sherlock", 24, "London, 221B Baker Street")
cat = Cat("Simon", 10, "british", owner)

Method Resolution Order (MRO)

Наследование — это очень мощный инструмент. Наследоваться можно не только от одного класса, а можно сразу от нескольких. Таким образом можно получать объекты, сочетающие в себе свойства многих классов. Тут должен возникнуть вопрос, а что будет, если у нескольких классов есть атрибуты с одинаковым именем?

Для ответа на этот вопрос надо понять, как Python ищет атрибуты (поля или методы) в объектах. Собственно, это и есть MRO (Method Resolution Order). MRO в Python работает следующим образом:

    Ищет атрибут среди аттрибутов самого класса. Именно благодаря этому, вы можете "переопределять" родительские атрибуты.
    Ищет атрибут у первого родителя (тот, что указан первым в списке родителей).
    Ищет атрибут у следующего родителя в списке родителей, пока такие есть.
    Ищет атрибут в родителях первого родителя.
    Повторяет п.4 для всех родителей.
    Вызывает исключение, что атрибут не найден.

Поиски заканчиваются, как только атрибут найден.

class A:
    x = 'I am A class'


class B:
    x = 'I am B class'
    y = 'I exist only in B'


class C(A, B):
    z = "This exists only in C"


c = C()
print(c.z)  # This exists only in C
print(c.y)  # I exist only in B
print(c.x)  # I am A class

Из этого примера видно, что в классе C поле x берется из A класса. Если же в этом же примере поменять список родителей, то получим:

class A:
    x = 'I am A class'


class B:
    x = 'I am B class'
    y = 'I exist only in B'


class C(B, A):
    z = "This exists only in C"


c = C()
print(c.z)  # This exists only in C
print(c.y)  # I exist only in B
print(c.x)  # I am B class

Теперь в классе C поле x берется из B класса.
Создайте два класса: CatDog и DogCat. Эти классы должны наследоваться от двух классов сразу: Cat и Dog. После наследования у экземпляра класса CatDog, родительский метод say должен возвращать "Meow", а у класса DogCat — "Woof". Для обоих указанных классов реализуйте метод info, который возвращает строку в следующем формате f"{self.nickname}-{self.weight}".

```python
class Animal:
    def __init__(self, nickname, weight):
        self.nickname = nickname
        self.weight = weight

    def say(self):
        pass


class Cat(Animal):
    def say(self):
        return "Meow"


class Dog(Animal):
    def say(self):
        return "Woof"


class CatDog(Cat, Dog):
    def info(self):
        return f"{self.nickname}-{self.weight}"


class DogCat(Dog, Cat):
    def info(self):
        return f"{self.nickname}-{self.weight}"
```

Контейнеры, созданные при помощи наследования

Часто для работы нужно создать объекты, которые ведут себя как стандартные контейнеры Python, но с модифицированным поведением. Вы конечно можете попробовать наследовать dict, str, list, но это может привести к ряду непредвиденных ошибок. Правильный способ получить модифицированный контейнер — это использовать пакет collections и классы UserList, UserDict, UserString, которые в нем есть.

Все эти классы ведут себя в точности как встроенные контейнеры с тем лишь отличием, что сами данные лежат в поле data у этих классов и вы можете использовать это поле по своему усмотрению.

from collections import UserDict


class ValueSearchableDict(UserDict):
    def has_in_values(self, value):
        return value in self.data.values()


as_dict = ValueSearchableDict()
as_dict['a'] = 1
as_dict.has_in_values(1)  # True
as_dict.has_in_values(2)  # False

В этом примере мы создали класс, который ведет себя как словарь, но в нем есть дополнительный метод, который проверяет наличие некоторого value среди значений в этом словаре.

from collections import UserList


class CountableList(UserList):
    def sum(self):
        return sum(map(lambda x: int(x), self.data))


countable = CountableList([1, '2', 3, '4'])
countable.append('5')
countable.sum()  # 15

В этом примере мы создали класс, который ведет себя как список, но в нем есть метод sum, который возвращает сумму всего содержимого этого класса, при этом преобразуя строки в целые числа.

from collections import UserString


class TruncatedString(UserString):
    MAX_LEN = 7

    def truncate(self):
        self.data = self.data[:self.MAX_LEN]


ts = TruncatedString('abcdefghjklmnop')
ts.truncate()
print(ts)  # abcdefg

Последний пример показывает модифицированную сроку с методом truncate, который ограничивает размер строки до MAX_LEN символов.

В четвертом модуле мы реализовали функцию lookup_key для поиска всех ключей по значению в словаре. Первым параметром в функцию мы передавали словарь, а вторым — значение, которое хотели найти. Результатом был список ключей или пустой список, если мы ничего не находили.

def lookup_key(data, value):
    keys = []
    for key in data:
        if data[key] == value:
            keys.append(key)
    return keys

Создайте класс LookUpKeyDict, родителем которого будет класс UserDict. Сделайте функцию lookup_key методом класса LookUpKeyDict.

```python
from collections import UserDict


class LookUpKeyDict(UserDict):
    def lookup_key(self, value):
        keys = []
        for key in self.data:
            if self.data[key] == value:
                keys.append(key)
        return keys
```

Перепишем задачу расчета задолженностей по коммунальным услугам с помощью класса UserList.

payment = [1, -3, 4]


def amount_payment(payment):
    sum = 0
    for value in payment:
        if value > 0:
            sum = sum + value
    return sum

Напомним условие. У нас есть список показаний задолженностей по коммунальным услугам в конце месяца — список payment. Задолженности могут быть отрицательными — у нас переплата, или положительными, если необходимо оплатить по счетам.

Создайте класс AmountPaymentList, наследуйте его от класса UserList. Сделайте функцию amount_payment методом класса AmountPaymentList.

```python
from collections import UserList

payment = [1, -3, 4]


class AmountPaymentList(UserList):
    def amount_payment(self):
        sum = 0
        for self.value in payment:
            if self.value > 0:
                sum = sum + self.value
        return sum
```
